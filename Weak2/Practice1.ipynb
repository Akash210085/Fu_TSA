```python
import matplotlib.pyplot as plt

def plot_df(df,col):
    
    columns=df.loc[:,col:].columns.tolist()
    size=len(columns)
    fig,ax=plt.subplots(size,1, sharex=True)
    for i,colname in enumerate(columns):
        df[colname].plot(ax=ax[i],figsize=(16,10))
        ax[i].set_title(colname)
    plt.show()
```

```python
plot_df(df,'births')
```
